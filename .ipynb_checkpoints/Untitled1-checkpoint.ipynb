```python
import requests
from bs4 import BeautifulSoup

```

```python
def get_amazon_reviews(product_url, num_pages=2):
    reviews = []
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }

    for page in range(1, num_pages + 1):
        page_url = f"{product_url}&pageNumber={page}"
        response = requests.get(page_url, headers=headers)
        if response.status_code != 200:
            print("")
            continue

        soup = BeautifulSoup(response.content, 'html.parser')
        review_divs = soup.find_all('div', {'data-hook': 'review'})

        for review_div in review_divs:
            review_body = review_div.find('span', {'data-hook': 'review-body'})
            if review_body:
                review_text = review_body.get_text().strip()
                reviews.append(review_text)

    return reviews

# Example usage
product_url = 'https://www.amazon.in/ASUS-Zenbook-i5-12500H-Fingerprint-UX5401ZA-KM541WS/product-reviews/B0C5N4MM4M/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
reviews = get_amazon_reviews(product_url, num_pages=5)
for idx, review in enumerate(reviews):
    print(f"Review {idx+1}:\n{review}\n")
```

```python
import re
import string
import nltk
import pandas as pd
from nltk.corpus import stopwords
```

```python
def remove_emojis(text):
    return text.encode('ascii', 'ignore').decode('ascii')

def remove_punctuations(text):
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator)
```

```python
df = pd.DataFrame({'review': reviews})

df['lowercase'] = df['review'].apply(lambda x: ' '.join(word.lower() for word in x.split()))

df['remove_emojis'] = df['lowercase'].apply(remove_emojis)

df['remove_punctuations'] = df['remove_emojis'].apply(remove_punctuations)

stop_words = stopwords.words('english')

df['stopwords'] = df['remove_punctuations'].apply(lambda x: ' '.join(word for word in x.split() if word not in stop_words))

df['clean_review'] = df['stopwords']

df.head()
```

```python
# Calculate word count
df['word_count'] = df['stopwords'].apply(lambda x: len(x.split()))

# Calculate character count
df['char_count'] = df['remove_punctuations'].apply(lambda x: len(x))

# Calculate average word length
df['average_word_length'] = df.apply(lambda row: row['char_count'] / row['word_count'], axis=1)

# Calculate stopword count
df['stopword_count'] = df['remove_punctuations'].apply(lambda x: len([word for word in x.split() if word in stop_words]))

# Calculate stopword rate
df['stopword_rate'] = df['stopword_count'] / df['word_count']
```
