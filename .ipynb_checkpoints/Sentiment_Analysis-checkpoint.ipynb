```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from textblob import TextBlob
from wordcloud import WordCloud
import matplotlib.pyplot as plt
#nltk.download('wordnet')
```

```python
def remove_emojis(text):
    return text.encode('ascii', 'ignore').decode('ascii')

def remove_punctuations(text):
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator)
```

```python
# Function to get reviews from Amazon
def get_amazon_reviews(product_url, num_pages=2):
    reviews = []
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }

    for page in range(1, num_pages + 1):
        page_url = f"{product_url}&pageNumber={page}"
        response = requests.get(page_url, headers=headers)
        if response.status_code != 200:
            print(f"Failed to retrieve page {page}")
            continue

        soup = BeautifulSoup(response.content, 'html.parser')
        review_divs = soup.find_all('div', {'data-hook': 'review'})

        for review_div in review_divs:
            review_body = review_div.find('span', {'data-hook': 'review-body'})
            if review_body:
                review_text = review_body.get_text().strip()
                reviews.append(review_text)

    return reviews

# Example usage
product_url = 'https://www.amazon.in/ASUS-Zenbook-i5-12500H-Fingerprint-UX5401ZA-KM541WS/product-reviews/B0C5N4MM4M/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
reviews = get_amazon_reviews(product_url, num_pages=5)


```

```python
df = pd.DataFrame({'review': reviews})

# Apply lowercase transformation to a new column
df['lowercase'] = df['review'].apply(lambda x: ' '.join(word.lower() for word in x.split()))

df['remove_emojis'] = df['lowercase'].apply(remove_emojis)

# Remove punctuations from the text without emojis
df['remove_punctuations'] = df['remove_emojis'].apply(remove_punctuations)

stop_words = stopwords.words('english')

# Remove stopwords from the cleaned text
df['stopwords'] = df['remove_punctuations'].apply(lambda x: ' '.join(word for word in x.split() if word not in stop_words))

df['clean_review'] = df['stopwords']

# Calculate word count
df['word_count'] = df['stopwords'].apply(lambda x: len(x.split()))

# Calculate character count
df['char_count'] = df['remove_punctuations'].apply(lambda x: len(x))

# Calculate average word length
df['average_word_length'] = df.apply(lambda row: row['char_count'] / row['word_count'], axis=1)

# Calculate stopword count
df['stopword_count'] = df['remove_punctuations'].apply(lambda x: len([word for word in x.split() if word in stop_words]))

# Calculate stopword rate
df['stopword_rate'] = df['stopword_count'] / df['word_count']

lemmatizer = WordNetLemmatizer()
df['lema_clean_review'] = df['remove_punctuations'].apply(lambda x: ' '.join(lemmatizer.lemmatize(word) for word in word_tokenize(x) if word not in stop_words))


# Print the DataFrame
print(df)
```

```python
df['polarity'] = df['lema_clean_review'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['subjectivity'] = df['lema_clean_review'].apply(lambda x: TextBlob(x).sentiment.subjectivity)
df.describe()
```

```python
text = ' '.join(df['lema_clean_review'])

# Generate a WordCloud object
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Plot the WordCloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
def getAnalysis(score):
    if score < 0:
        return 'Negative'
    elif score == 0:
        return 'Neutral'
    else:
        return 'Positive'

# Apply the function to the 'polarity' column and create a new column 'Analysis'
df['Analysis'] = df['polarity'].apply(getAnalysis)
```

```python
df.drop(['lowercase', 'remove_emojis', 'remove_punctuations', 'stopwords', 'clean_review', 'word_count', 'char_count', 'average_word_length', 'stopword_count', 'stopword_rate', 'lema_clean_review'], axis=1, inplace=True)
```

```python
# Plotting Polarity & Subjectivity
plt.figure(figsize=(10, 6))
plt.scatter(df['polarity'], df['subjectivity'], color='blue')
plt.title('Sentiment Analysis')
plt.xlabel('Polarity')
plt.ylabel('Subjectivity')
plt.show()
```

```python
positive_reviews = df[df['Analysis'] == 'Positive'].shape[0]
negative_reviews = df[df['Analysis'] == 'Negative'].shape[0]
total_reviews = df.shape[0]

positive_percentage = (positive_reviews / total_reviews) * 100
negative_percentage = (negative_reviews / total_reviews) * 100

print(f"Percentage of Positive Reviews: {positive_percentage:.2f}%")
print(f"Percentage of Negative Reviews: {negative_percentage:.2f}%")
```

```python
counts = [positive_reviews, negative_reviews]
labels = ['Positive', 'Negative']

# Plotting the bar chart
plt.figure(figsize=(8, 6))
plt.bar(labels, counts, color=['green', 'red'])
plt.title('Count of Positive and Negative Reviews')
plt.xlabel('Sentiment')
plt.ylabel('Number of Reviews')
plt.show()
```
